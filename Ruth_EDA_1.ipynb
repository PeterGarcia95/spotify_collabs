# Spotify Collaborations Dash EDA 

```python
# Load relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_us = pd.read_csv("2021.01.19 spotify_us_all_fe.csv")
top_us.head()
```

### Top Month by Average Streams

```python
top_month_str = top_us.groupby('Album_release_month')['Streams'].mean().reset_index().sort_values(by='Streams', ascending = False).reset_index() 

top_month_str
```

```python
top_month_str = top_us.groupby('Album_release_month')['Streams'].mean().reset_index().sort_values(by='Streams', ascending = False).reset_index().plot.bar(x= "Album_release_month", y = "Streams", color = 'limegreen' )  


top_month_str.set_xlabel("Month")
top_month_str.set_ylabel("Streams")
top_month_str.set_facecolor('black')
top_month_str.set_title("Songs Released in April Have Higher Streams On Average") 

```

### Top Day Of Week by Average Streams

```python
top_day_str = top_us.groupby('Album_release_dayweek')['Streams'].mean().reset_index().sort_values(by='Streams', ascending = False).reset_index() 

top_day_str
```

```python
top_day_str_chart = top_us.groupby('Album_release_dayweek')['Streams'].mean().reset_index().sort_values(by='Streams', ascending = False).reset_index().plot.bar(x= "Album_release_dayweek", y = "Streams", color = 'limegreen' )  

top_day_str_chart.set_xlabel("Day of the Week")
top_day_str_chart.set_ylabel("Streams")
top_day_str_chart.set_facecolor('black')
top_day_str_chart.set_title("Songs Released on a Saturday Have Higher Streams On Average") 
```
