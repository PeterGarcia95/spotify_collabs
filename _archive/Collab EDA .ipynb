## Variables Outline 
1- Date: The date a given song appears on the Top 200 chart.

2- Track URL: Link to the Spotify page for the song.

3- Position: Ranking according to number of streams in a region.

4- Track Name: Name of the track/song.

5- Artist: Name of the main artist.

6- Streams: Number of streams on a given Date.

7- Country: The country of the chart.

8- Track URI: Spotify's unique identifier for the song.

9- Collaborator URI: Spotify's unique identifier for the artist(s).

10- No. of Collaborators: Number of distinct artists credited for a given song.

11- Song_min_date: The earliest day a given song appears in a given country's chart.

12- Song_days_since_first: The number of days elapsed between Song_min_date and the current Date.

13- Song_days_onchart: The number of days the song has appeared on the chart in the given country.

14- Artist_min_date: The earliest day a given artist appears in a given country's chart.

15- Artist_new_song: Binary variable, equals 1 if this is the first time a given song from a given artist appears in a given country's chart.

16- Date_lag: Date, lagged by one row when grouped by Country, Artist, Track Name. Gives the date of the song's last appearance prior to the current Date.

17- Date_diff: The difference between the current Date and the Date_lag.

18- **Song_new_streak: Binary variable, equals 1 if this is the start of ??????

19- Song_streak_id: Unique identifier of a given song's streak in a given country.

20- Song_consec_day: Unique identifier of a day within a unique streak id.

21- Song_cumu_streams: The cumulative total number of streams for a given song in a given country up until the current Date. 

22- Song_cumu_mean_streams: The cumulative average number of streams for a given song in a given country up until the current Date. Calculated as Song_cumu_streams / Song_days_onchart.

23- Artist_cumsum_songs: The given artist's cumulative total number of songs up until the current Date.

24- Artist_cumu_songs: The given artist's cumulative total number of songs up until the current Date.

25- Artist_days_onchart: The given artist's cumulative number of days/appearance on a given country's chart up until the current Date.

26- Artist_cumu_streams: The given artist's cumulative total number of streams from all their songs in a given country's chart up until the current Date.

27- Week_id: The ID of the week within a year, resets on a Monday (this week ID will reset back to 0 when a new year begins)

28- Year_week_id: A concatenation of the week ID and year

29- Day_week: Day of the week, in 3-letter abbreviation

30- Week_id_BB: The ID of the week within a year, but resets on a Tuesday to match with Billboard's week.

31- Week_mean_streams: The mean number of streams for a given Year_week_id for a given country.

## Inspecting songs' position against variables of interest


````python
# graph average streams vs Position

ax = (
    top_us.groupby(["Streams"])["Position"]
    .mean()
    .reset_index(name="Position")
    .sort_values(by="Streams", ascending= False)
    .plot.scatter(x="Streams", y="Position")
)
ax.set_xlabel("Streams (in millions)")
ax.set_ylabel("Average Position")
a = ax.set_title("Songs with higher streams are ranked higher in chart")
````

````python
# graph song cumulative mean streams vs Position

ax1 = (
    top_us.groupby(["Song_cumu_mean_streams"])["Position"]
    .mean()
    .reset_index(name="Position")
    .sort_values(by="Position", ascending=False)
    .plot.scatter(x="Song_cumu_mean_streams", y="Position")
)
ax1.set_xlabel("Average song cumulative streams (in millions)")
ax1.set_ylabel("Average Position")
b = ax1.set_title('Songs with higher streams are ranked higher in chart')
````

````python
# change No. of Collaborators from float to integer 
top_us['No. of Collaborators']= top_us['No. of Collaborators'].astype(int)
````

````python
# graph No of collaborators vs Position
ax2 = (
    top_us.groupby(["No. of Collaborators"])["Position"]
    .mean()
    .reset_index(name="Position")
    .sort_values(by="No. of Collaborators", ascending= True)
    .plot.bar(x="No. of Collaborators", y="Position")
)
ax2.set_xlabel("No. of Collaborators")
ax2.set_ylabel("Average Position")
c = ax2.set_title("The position of a song drops with more collaborators up to 8 artists")
````

````python
# boxplot graph for No of collaborators vs Position 
sns.boxplot(x = 'No. of Collaborators', y='Position', data = top_us)
plt.ylabel("Position")
plt.title("The position of a song drops with more collaborators up to 8 artists")
````

````python
# graph days a song is on chart vs average position
ax3 = (
    top_us.groupby(["Song_days_onchart"])["Position"]
    .mean()
    .reset_index(name="Position")
    .sort_values(by="Song_days_onchart", ascending= True)
    .plot.line(x="Song_days_onchart", y="Position")
)
ax3.set_xlabel("Song_days_onchart")
ax3.set_ylabel("Average Position")
d = ax3.set_title("Initial increase in song ranking with more days on chart, followed by a drop in position post ~ 100 days") 
````

````python
top_10=top_us[top_us['Position']<= 10]
bottom_10 = top_us[top_us['Position']>= 190]
````

````python
# graph days a song is on chart vs average position for songs ranked in top 10 
ax4 = (
    top_10.groupby(["Song_days_onchart"])["Position"]
    .mean()
    .reset_index(name="Position")
    .sort_values(by="Song_days_onchart", ascending= True)
    .plot.line(x="Song_days_onchart", y="Position")
)
ax4.set_xlabel("Song_days_onchart")
ax4.set_ylabel("Average Position")
e = ax4.set_title("Songs's position vs # of days on chart for songs ranked in top 10 in chart") 
````

````python
# graph days a song is on chart vs average position for songs ranked in bottom 10 

ax5 = (
    bottom_10.groupby(["Song_days_onchart"])["Position"]
    .mean()
    .reset_index(name="Position")
    .sort_values(by="Song_days_onchart", ascending= True)
    .plot.line(x="Song_days_onchart", y="Position")
)
ax5.set_xlabel("Song_days_onchart")
ax5.set_ylabel("Average Position")
f = ax5.set_title("Songs's position vs # of days on chart for songs ranked in bottom 10 in chart") 
````

````python
# plotting Artist_days_onchart vs average position 
ax6 = (
    top_us.groupby(["Artist_days_onchart"])["Position"]
    .mean()
    .reset_index(name="Position")
    .sort_values(by="Position", ascending=True)
    .plot.hist(x="Artist_days_onchart", y="Position")
)
ax6.set_xlabel("Artist's average days on chart")
ax6.set_ylabel("Average Position")
g = ax6.set_title("Artist's average days on chart vs average position ")
````

````python
# plotting song_new_streak vs average position 
ax7 = (
    top_us.groupby(["Song_new_streak"])["Position"]
    .mean()
    .reset_index(name="Position")
    .sort_values(by="Song_new_streak", ascending=True)
    .plot.bar(x="Song_new_streak", y="Position")
)
ax7.set_xlabel("Song new streak")
ax7.set_ylabel("Average Position")
h = ax7.set_title("Song new streak vs average position")
````

````python
sns.boxplot(x = 'Song_new_streak', y='Position', data = top_us)
plt.ylabel("Position")
plt.title("boxplot of song new streak vs average position") 
````
